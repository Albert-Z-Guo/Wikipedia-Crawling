```python
# method to remove all section titles
input_directory_name = 'people_won_awards_1910_2010_en'
# output_directory_name = 'people_won_awards_1910_2010_en_processed'

start = 0
num_display = 1

filenames = os.listdir(input_directory_name)
for n in range(start, len(filenames)):
    file_path = '{}/{}'.format(input_directory_name, filenames[n])
    with open(file_path, 'r') as file:
        content = file.read()
        print(content, end='\n\nAfter preprocessing:\n\n')
        
        # remove section titles
        pattern = re.compile('==+ .+ ==+')
        sections = pattern.split(content)
        sections = [i.strip() for i in sections if len(i.strip()) != 0]
        print('\n\n'.join(sections))
    n += 1
    if n == start + num_display:
        break
    print('\n\n----------\n\n')
```
